```python
import pandas as pd
from pandas.core.interchange.dataframe_protocol import DataFrame

df_core = pd.read_csv('ohchr_instruments-core.csv')
df_universal = pd.read_csv('ohchr_instruments-universal.csv')
```

```python
df_core.tail()
```

```python
df_universal.tail()
```

```python
print('Number of instruments in core', len(df_core.index))
print('Number of instruments in universal', len(df_universal.index))

print('Core instruments in the universal instruments: ', len(df_universal[df_universal['title'].isin(df_core['title'])]))

print('Universal instruments in the core instruments: ', len(df_core[df_core['title'].isin(df_universal['title'])]))
```

```python
df = pd.concat([df_core, df_universal])
df.drop_duplicates(subset='title', keep='last', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# Drop the 'page' column
if 'page' in df.columns:
    df.drop(columns=['page'], inplace=True)

df.to_csv('./ohchr_instruments.csv', index=False)
```
